```python
import math
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
class trig_fn:
  # 매게변수 st, ed, n을 입력받아옴
  def setting(self, st, ed, n):
    self.st = st
    self.ed = ed
    self.n = n
  ####### 여러분이 작성하세요. #######
  # 문제의 요구조건(2)를 수행하는 generate_x_y 함수를 만드세요.
  def generate_x_y(self, opt):
    self.x = []
    self.y = []
    self.dx = (self.ed-self.st)/(self.n-1)
    for i in range(self.n):
      x_ = self.st + self.dx*i
      self.x.append(x_)
      if opt == 1:
        self.y.append(math.sin(x_))
        # 이 부분에 작성
      elif opt == 2:
        self.y.append(math.cos(x_))
        # 이 부분에 작성
      elif opt == 3:
        self.y.append(math.tan(x_))
        # 이 부분에 작성
      else: opt == 1000000
        # 이 부분에 작성
  ###################################
  x=np.linspace(0,2*np.pi,100)



  # 결과를 출력함
  def plot(self):
    plt.plot(self.x,self.y)
    plt.show()
```

```python
g1=trig_fn()
g1.setting(0, 2*math.pi, 100)
g1.generate_x_y(1)
g1.plot()
g2=trig_fn()
g2.setting(0, 2*math.pi, 100)
g2.generate_x_y(2)
g2.plot()
g3=trig_fn()
g3.setting(0, 2*math.pi, 100)
g3.generate_x_y(3)
g3.plot()
```
